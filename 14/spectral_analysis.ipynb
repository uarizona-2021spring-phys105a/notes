# Spectral Analysis of Different VLBI Closure Quantities

This Jupyter Notebook explores using simple spectral analysis to study VLBI closure quantities

We start by loading the necessary libraries; define functions to generate and manipulate time series and spectra

```python
%load_ext autoreload
%autoreload 2

import numpy             as np
import scipy.signal      as ss

import matplotlib.pyplot as plt
%matplotlib inline

from spectral_analysis_helpers import * # non-essential helper functions
```

```python
def rand_complex(N):
    return (np.random.normal(size=N) +
            np.random.normal(size=N) * 1j)

def spec_filter(ts, filter, mean=None, **kwargs):
    k = np.fft.fftfreq(len(ts))
    s = np.fft.fft(ts)
    f = np.concatenate(([1.0], # filtering should not affect the mean
                        np.sqrt(filter(k[1:], **kwargs))))
    g = f * s
    if mean is not None:
        g[0] = mean # specify the mean if it is set
    return np.fft.ifft(g)

def varnorm(ts, mean=None):
    m  = np.mean(ts)
    d  = ts - m
    d /= np.sqrt(np.mean(abs(d)**2))
    if mean is None:
        return d + m # preserve the mean of the time series
    else:
        return d + mean
    
def ampclip(ts, min=0.8, max=1.0):
    return np.clip(np.absolute(ts), min, max) * np.exp(np.angle(ts)*1j)

def Gaussian(k, sigma=1.0):
    return np.exp(-0.5 * (k/sigma)**2)

def powerlaw(k, power=-2.0):
    return np.power(abs(k), power)
```

## Basic Time Series Examples

Generate and plot different time series and spectra.  Note that the spectrum of the Gaussian filtered time series has a power law tail.  This is because we used Bartlett window to compute/estimeate the power spectra.

```python
n = np.real(varnorm(rand_complex(16384), mean=0.0))
G = np.real(varnorm(spec_filter(n, Gaussian, sigma=16/len(n))))
p = np.real(varnorm(spec_filter(n, powerlaw, power=-2.0)))

f, ax = plt.subplots(2,3,figsize=(18,12))
plot_ts_spec(ax, 0, n, title="White Noise")
plot_ts_spec(ax, 1, G, title="Time series with Gaussian spectrum")
plot_ts_spec(ax, 2, p, title="Time series with power law spectrum")
plt.show(f)
```

Filtering can remove random noise if done right.  Otherwise, residual noise can remain or the signal itself can be filtered out

```python
snr = 1.0
ts  = G + n / np.sqrt(snr)

f, ax = plt.subplots(2,4,figsize=(24,12))
plot_ts_spec(ax, 0, ts,                  title="Time Series = signal + noise"); plot_ts_spec(ax, 0, G)
plot_ts_spec(ax, 1, ss.medfilt(ts,  21), title="Median filter with width=21" ); plot_ts_spec(ax, 1, G)
plot_ts_spec(ax, 2, ss.medfilt(ts, 101), title="Median filter with width=101"); plot_ts_spec(ax, 2, G)
plot_ts_spec(ax, 3, ss.medfilt(ts, 501), title="Median filter with width=501"); plot_ts_spec(ax, 3, G)
plt.show(f)
```

## Basic VLBI Specific Examples

For VLBI application, we model the visbility between stations 1 and 2, `vis12`, by a Gaussian filtered random signal.  We can apply two gains, `gain1` and `gain2`, and random noise `noise12` to it.  The gains are obtained by amplitude clipping Gaussian filtered random signals.

If the application of gains were addiative, there would have been three features seen in the power spectrum, namely, the visibility, the gains, and the random noise.  This is shown in the second column of the plot.

However, in reality, the gains are multiplictive.  The spectrum of the visiblity and gains is a *convolution* of the spectra of the visibility and the gains.  The component that has the wider spectrum wins so only the gain spectrum (and the thermal noise) surives in the third column.

In the four column, we plot the amplitude only and its power spectrum.  It is interesting to see the original signal seems to show up again.  A hand wavy argument goes as the following: the transformation of a complex variable $z = x + iy \mapsto r + i\theta$ is done by the exponential map.  All information that's not encoded in the phase is naturally encoded in $r = \ln|z|$.  Therefore, multiplicative operations show up addictive in when we rewrite a complex variable in terms of amplitude and phase.  (Credit: Lindy Blackburn brought to my attention that $\ln|\,\cdot\,|$ is a more natural quantitiy to look at compare to $|\,\cdot\,|$.)

To verify the above argument, we plot $\ln|\,\cdot\,|$ of different components of the signal in the last column.  The thermal noise spectrum is totally off because of its addicitive natural.  Nevertheless, the spectrum of the time series traces both the visibility and gains spectra.

P.S. This multiplitive--additivie transform in the amplitude is very interesting.  It may allow us to develop filtering technique to remove the gain and leakage *without* actually solving for them.  I forsee its potential in improving calibration.  (P.S. Maciej Wielgus told me that this is called cepstrum.)

```python
N   = 32 * 8 * 4096
snr = 1000.0

vis12   =         varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32  /N))
gain1   = ampclip(varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32*8/N)))
gain2   = ampclip(varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32*8/N)))
noise12 =                             rand_complex(N)

ts_add = vis12 + gain1 * np.conj(gain2) + noise12 / np.sqrt(snr)
ts_mul = vis12 * gain1 * np.conj(gain2) + noise12 / np.sqrt(snr)

f, ax = plt.subplots(2,5,figsize=(30,12))
ts = vis12;                                       plot_ts_spec(ax, 0, ts)
ts = gain1 * np.conj(gain2);                      plot_ts_spec(ax, 0, ts)
ts = noise12 / np.sqrt(snr);                      plot_ts_spec(ax, 0, ts, title="Original components")
ts = ts_add;                                      plot_ts_spec(ax, 1, ts, title="If gains were addictive")
ts = ts_mul;                                      plot_ts_spec(ax, 2, ts, title="Multiplicative gains")
ts = np.absolute(ts_mul);                         plot_ts_spec(ax, 3, ts, title="Amplitude Only")
ts = np.log(np.absolute(vis12));                  plot_ts_spec(ax, 4, ts)
ts = np.log(np.absolute(gain1 * np.conj(gain2))); plot_ts_spec(ax, 4, ts)
ts = np.log(np.absolute(noise12 / np.sqrt(snr))); plot_ts_spec(ax, 4, ts)
ts = np.log(np.absolute(ts_mul));                 plot_ts_spec(ax, 4, ts, title="ln(Amplitude) of components")
plt.show(f)
```

The relations between different curves are more clear when we overplot the first column with the 3rd and 4th columns.

```python
f, ax = plt.subplots(1, 2, figsize=(16,8))
ts = vis12;                                                   plot_specs(ax, ts, ["Ideal Visibility",   "Amplitude of Ideal Vis."])
ts = gain1 * np.conj(gain2);                                  plot_specs(ax, ts, ["Gain1*~Gain2",       "Amplitude of Gain1*~Gain2"])
ts = noise12 / np.sqrt(snr);                                  plot_specs(ax, ts, ["Noise",              "Amplitude of Noise"])
ts = vis12 * gain1 * np.conj(gain2) + noise12 / np.sqrt(snr); plot_specs(ax, ts, ["Observed Visibility","Amplitude of Obs. Vis."])
plt.show(f)
```

## Minic VLBI Observation with Nearby Stations

We create four visibilities to minic VLBI observation.  We assume that stations 1 and 2 are nearby, and stations 3 and 4 are nearby.  Because `vis12` and `vis34` are on short baselines, we assume them vary slowly.  Also, we assume stations 1 and 2, and tations 3 and 4, have the same atmospheres, respecitvely.  This allows us to set `gain2` = `gain1` and `gain4` = `gain3`.

```python
N   = 32 * 8 * 4096
snr = 1000.0

vis12 = varnorm(spec_filter(rand_complex(N), Gaussian, sigma=1 /len(ts)))/2 + 1
vis23 = varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32/len(ts)))
vis34 = varnorm(spec_filter(rand_complex(N), Gaussian, sigma=1 /len(ts)))/2 + 1
vis41 = varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32/len(ts)))

gain1 = ampclip(varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32*8/len(ts))))
gain2 = gain1
gain3 = ampclip(varnorm(spec_filter(rand_complex(N), Gaussian, sigma=32*8/len(ts))))
gain4 = gain3

noise12 = rand_complex(N)
noise23 = rand_complex(N)
noise34 = rand_complex(N)
noise41 = rand_complex(N)

v12 = vis12 * gain1 * np.conj(gain2) + noise12 / np.sqrt(snr)
v23 = vis23 * gain2 * np.conj(gain3) + noise23 / np.sqrt(snr)
v34 = vis34 * gain3 * np.conj(gain4) + noise34 / np.sqrt(snr)
v41 = vis41 * gain4 * np.conj(gain1) + noise41 / np.sqrt(snr)
```

The following plots are busy.  The first two panels are the same except that the first panel uses the full compmlex visiblity, while the second panel uses only the visibility amplitude.  The most interesting features are:

- In the first panel, the spectra of the observed visibilities (purple and brown) trace the gain spectra (green and red).

- The same is true in the second panel but the amplitude of observed visibility 2--3 (brown) also picks up the ideal visibility spectra (orange).

- In both case, no gain is seen in the closure amplitudes (pink).  The spectra of the closure amplitudes is composed by the spectra of the ideal visibilities and thermal noise.

- Depending on how the data is being processed, we may not be able to fit for the thermal noise in the observed complex visibilities.  But we may be able to do it for the closure amplitude.  It will be nice to derive a relationship between these two different noises.  (P.S. the noise in closure amplitude is probably dominated by the sharp peaks caused by small denominator.  Maybe we can just mask them out?)

- The spectral shape of the ideal visbilities and the closure amplitude look the same.

The rightmost panel summary the most important spectra of the amplitudes.

```python
f, ax = plt.subplots(1, 3, figsize=(24,8))

plot_specs(ax, vis12,                     ["Ideal Visibility12",    "Amplitude of Ideal Vis.12"])
plot_specs(ax, vis23,                     ["Ideal Visibility23",    "Amplitude of Ideal Vis.23"])
plot_specs(ax, gain1 * np.conj(gain2),    ["Gain1*~Gain2",          "Amplitude of Gain1*~Gain2"])
plot_specs(ax, gain2 * np.conj(gain3),    ["Gain2*~Gain3",          "Amplitude of Gain2*~Gain3"])
plot_specs(ax, v12,                       ["Observed Visibility12", "Amplitude of Obs. Vis.12" ])
plot_specs(ax, v23,                       ["Observed Visibility23", "Amplitude of Obs. Vis.23" ])
plot_specs(ax, (v23 * v41) / (v12 * v34), ["Closure Visbility",     "Closure Amplitude"        ])

ts = vis23;                     ax[2].loglog(powerspec(abs(ts), len(ts)//32), label='Amplitude of Ideal Vis.23')
ts = v23;                       ax[2].loglog(powerspec(abs(ts), len(ts)//32), label='Amplitude of Obs. Vis.23')
ts = (v23 * v41) / (v12 * v34); ax[2].loglog(powerspec(abs(ts), len(ts)//32), label='Closure Amplitude')

ax[2].set_ylim((1e-3,1e9))
ax[2].set_xlabel('Frequency')
ax[2].legend()

plt.show(f)
```
